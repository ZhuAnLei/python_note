```python
predictor = dlib.shape_predictor(r'D:\A_TargetDetection\AnLee_Code\Mediapipe_FaceMesh\data\shape_predictor_68_face_landmarks.dat')
```

```python
#摄像头人眼ear检测
# 处理函数
def process_frame(img):
    #记录该帧开始处理的时间
    start_time = time.time()
    scaler = 1 #文字大小
    #BGR转RGB
    img_RGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    # 将RGB图像输入模型 获取预测结果
    # 使用特征提取器get_frontal_face_detector
    detector = dlib.get_frontal_face_detector()
    dic = [37,38,41,40,36,39]
    dets = detector(img_RGB, 1)
    lefteye = []
    if len(dets) == 0:
        img = cv2.putText(img_RGB, "No body face detected!", (25*scaler,150*scaler), cv2.FONT_HERSHEY_DUPLEX, 0.8, (0, 255, 255), 2, cv2.LINE_AA)
    else:
        for det in dets:
            shape = predictor(img_RGB, det)
            for i in dic:
                lefteye.append(shape.part(i))
                cv2.putText(img_RGB, str(i), (shape.part(i).x, shape.part(i).y), cv2.FONT_HERSHEY_DUPLEX, 0.8, (0, 255, 255), 2, cv2.LINE_AA)
                cv2.circle(img_RGB, (shape.part(i).x, shape.part(i).y), 1, (0, 0, 255))

        # point 转 array
        point_arr = []
        for point in lefteye:
            point_arr.append([point.x,point.y])
        
        point_arr = np.array(point_arr)
        A = np.linalg.norm(point_arr[0]-point_arr[2])
        B = np.linalg.norm(point_arr[1]-point_arr[3])
        C = np.linalg.norm(point_arr[4]-point_arr[5])
        ear = (A + B) / (2.0 * C) 
        ear = round(ear,2)
        print(ear)
        cv2.putText(img_RGB, "Left Eye Aspect Ratio:{}".format(str(ear)), (25*scaler,150*scaler), cv2.FONT_HERSHEY_DUPLEX, 2, (0, 255, 255), 2, cv2.LINE_AA)

    #记录该帧处理完毕时间
    end_time = time.time()
    # 计算每秒处理图像帧数FPS
    FPS = 1/(end_time-start_time)

    img = cv2.putText(img,'FPS  '+str(int(FPS)),(25*scaler,100*scaler),cv2.FONT_HERSHEY_SIMPLEX,1.25*scaler,(255,0,255),2*scaler)

    return img
```

```python
import cv2
import time
cap = cv2.VideoCapture(0)
cap.open(0)
#无限循环，知道break被触发
while cap.isOpened():
    success ,frame = cap.read()
    if not success:
        print('ERROR')
        break

    ### 处理帧函数
    frame = process_frame(frame)

    #展示处理后的三通道图像
    cv2.imshow('my_window',frame)
    if cv2.waitKey(1) in [ord('q'),27]: # 按键盘上的q或者esc退出（在英文输入法下）
        break

# 关闭摄像头
cap.release()

# 关闭图像窗口
cv2.destroyAllWindows()
```
